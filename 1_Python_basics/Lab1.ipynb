```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
irisRaw = load_iris()
```

```python
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
#1.
iris
```

```python
#2.
print(f"Ilość wierszy: {iris.shape[0]}")
print(f"Ilość kolumn: {iris.shape[1]}")
```

```python
#3.
iris.describe()
```

```python
#4.
iris.groupby(['target']).mean()
```

```python
#6.
iris.dropna(axis=0, how='any', inplace=True)
iris.dropna(axis=1, how='any', inplace=True)
iris
```

```python
#7.
iris.sort_values(by=['sepal width (cm)'], inplace=True)
iris
```

```python
#8.
min = iris.min()['petal length (cm)']
max = iris.max()['petal length (cm)']
print(f'Minimalna długość płatka: {min}')
print(f'Maksymalna długość płatka: {max}')
```

```python
#9.
iris.std()
```

```python
#10.
iris2 = iris[iris['sepal length (cm)'] > iris.mean()['sepal length (cm)']]
iris2
```

```python
#11.
m = np.sort(pd.unique(iris['target']))
# iris['target'].value_counts()
x = np.zeros((50, 3))
for i in m:
    x[:,int(i)] = iris[iris['target']==i]['sepal length (cm)']

colors = ['red', 'tan', 'lime']
labels = ['Klasa 0', 'Klasa 1', 'Klasa 2']
plt.hist(x, bins=8, histtype='bar', color=colors, label=labels)
plt.title('Histogram dla wartości sepal length dla poszczególnych klas')
plt.xlabel('Wartość sepal length (cm)')
plt.ylabel('Ilość wystąpień w bazie danych')
plt.grid()
plt.legend()
plt.show()
```
